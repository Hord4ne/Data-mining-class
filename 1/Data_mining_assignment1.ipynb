```
random_uniform = np.random.uniform(-1,1,10000)
random_uniform.shape
```

```
def uniform_density(x,low,hi):
    return 1.0/(hi-low) if low <= x <= hi else 0
```

```
def plot_hist_density(hist_data,bins, plot_x, plot_y, distrubition_name):
    plt.figure(figsize=(14,6))

    plt.subplot(1,3,1)
    plt.hist(hist_data,bins = bins,edgecolor='black')
    plt.title('Histogram of '+ str(hist_data.shape[0]) + ' samples')
    plt.subplot(1,3,2)
    plt.plot(plot_x,plot_y,color='r')
    plt.title('Density function for '+ distrubition_name + ' distribution')
    
    plt.subplot(1,3,3)
    plt.hist(hist_data, density= True, bins = bins,edgecolor='black')
    plt.plot(plot_x,plot_y,color='r')
    plt.title('Combined')
    plt.show()
```

```
x = np.arange(-1.5,1.5,0.001)
y = [uniform_density(a,-1,1) for a in x]
plot_hist_density(random_uniform,100,x ,y, 'uniform')
```

```
random_uniform = np.random.uniform(-1,1,100000)
x = np.arange(-1.5,1.5,0.001)
y = [uniform_density(a,-1,1) for a in x]
plot_hist_density(random_uniform,100,x ,y, 'uniform')
```

```
stacked_points_of_cross = np.vstack((points_of_cross_x,points_of_cross_y))
stacked_points_of_cross
```

```
def get_points_of_cross_with_line(a, b, stacked_points):
    new_a = -1.0/a
    new_b = -new_a*stacked_points[0,:] + stacked_points[1,:]
    points_cross_x = (new_b - b)/(a+ 1.0/a )
    points_cross_y = a*points_cross_x+b
    return np.vstack((points_cross_x,points_cross_y))
```

```
get_points_of_cross_with_line(2,5,stacked)
```

```
def calc_dist(stacked_points_1, stacked_points_2):
    return np.sqrt(np.sum((stacked_points_1 - stacked_points_2)**2,axis=0))
```

```
calc_dist(stacked,stacked_points_of_cross)
```

```
def sum_of_dists(stacked_points_1,stacked_points_2):
    return np.sum(calc_dist(stacked_points_1, stacked_points_2))
```

```
sum_of_dists(stacked,stacked_points_of_cross)
```

```
def calc_sum_from_line_parameters(a,b,stacked_points):
    cross_points = get_points_of_cross_with_line(a, b, stacked_points)
    return sum_of_dists(stacked_points,cross_points)
```

```
calc_sum_from_line_parameters(2,5,stacked)
```

```
#niech wielokoaty maja jeden wierzcholek zawsze w 0,0
def get_vertexes(k,d):
    result = np.zeros((2,k+1))
    initial_alpha = 2*np.pi/k
    alpha = initial_alpha
    point_x = 0
    point_y = 0
    result[0,0] = point_x
    result[1,0] = point_y
    result[0,k] = point_x
    result[1,k] = point_y
    result
    for i in range(1,k):
        x_addition = np.cos(alpha)*d
        y_addition = np.sin(alpha)*d
        point_x += x_addition
        point_y += y_addition
        result[0,i] = point_x
        result[1,i] = point_y
        alpha += initial_alpha
    return result
```

```
def get_distinct_vertexes(k,d):
    return get_vertexes(k,d)[:,:k]
```

```
def plot_figure(k,d):
    polygon = get_vertexes(k,d)
    plt.figure(figsize=(8,8))
    plt.plot(polygon[0],polygon[1])
    plt.show()
```

```
def generate_sample(a,b,times=1000):
    return np.vstack((np.random.normal(a,1,1000),np.random.normal(b,1,1000)))
```

```
def task_6(k,d):
    points_centers = get_distinct_vertexes(k,d)
    random_clouds = np.zeros((k,2,1000))    
    for i in range(k):
        a = points_centers[0][i]
        b = points_centers[1][i]
        random_cloud = generate_sample(a,b)
        random_clouds[i] = random_cloud
    return random_clouds
```

```
def plot_task_6_results(results,figsize=(8,8)):
    plt.figure(figsize=figsize)
    plt.scatter(results[:,0,:],results[:,1,:])
    plt.show()
```

```
def get_distances_matrix(cloud,points):    
    return np.sqrt(-2*np.dot(cloud,points.T) + np.sum(points**2,axis=1) + np.sum(cloud**2,axis=1)[:,np.newaxis])
```

```
def get_colors(k,d,clouds):
    points_centers = get_distinct_vertexes(k,d)
    cols = np.zeros((k,1000))
    for i in range (k):
        dist_matrix = get_distances_matrix(clouds[i].T,points_centers.T)
        min_distances = np.argmin(dist_matrix,axis=1)
        closest_to_own_center = min_distances == i
        
        cols[i] = closest_to_own_center
    return cols
    
```

```
def task_7_with_plot(k,d,figsize=(8,8)):
    clouds = task_6(k,d)
    colors = get_colors(k,d,clouds)
    
    plt.figure(figsize=figsize)
    for i in range(k):
        plt.scatter(clouds[i,0,:],clouds[i,1,:],c=['b' if colors[i][q] else 'r' for q in range(1000)])
    #print(np.sum(colors==0))
    print("Percent of red dots = ",np.sum(colors==0)/(10.0*k))
    plt.show()
```
